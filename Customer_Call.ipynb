```python
import pandas as pd
df = pd.read_excel(r"C:\Users\kevin\Downloads\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
#df = df.drop(columns="Not_Useful_Column")
df.head()
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("/..._123")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df.head()
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df.head()
```

```python
df["Phone_Number"] = df["Phone_Number"].replace('nan----','')
df["Phone_Number"] = df["Phone_Number"].replace('Na----','')
df
```

```python
df["Address"].str.split(',',n=2,expand=True)
```

```python
#df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',',n=2,expand=True)
#df = df.drop(columns=["Address"])
df
```

```python
#df["Paying Customer"] = df["Paying Customer"].replace('Y','Yes')
#df["Paying Customer"] = df["Paying Customer"].replace('N','No')
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('Y','Yes')
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('N','No')
df.fillna('', inplace=True)
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('','Yes')
df.head()
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Yes':
        df.drop(x, inplace = True)
df.head()
```

```python
df["Phone_Number"] = df["Phone_Number"].replace('','Yes')
df.head()
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == 'Yes':
        df.drop(x, inplace = True)
df.head()
```
